```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
```

```python
# define system in terms of a Numpy array
q=10
r=0.9
k1=0.8
k2=0.7
alpha=0.3
beta=0.1
s=0.8
d1=0.02
d2=0.03
def Sys(X, t=0):
    """ The rate equations that describe the BRD switch, """
    # here X[0] = [A] and x[1] = [B]    
    return np.array([ q+r*X[0]*(1-X[0]/k1)-alpha*X[0]*X[1] ,beta*X[1]*X[2]-d1*X[1],s*X[2]*(1-X[2]/k2)-beta*X[1]*X[2]-d2*X[2]])

# generate 1000 linearly spaced numbers for x-axes
t = np.linspace(0, 1000,  10000)
# initial values: A0 = 10, B0 = 2
Sys0 = np.array([3, 2,1])

# type "help(integrate.odeint)" if you want more information about integrate.odeint inputs and outputs.
X, infodict = integrate.odeint(Sys, Sys0, t, full_output=True)
# infodict['message']                      # integration successful

x,y,z = X.T

#plot
fig = plt.figure(figsize=(15,5))
fig.subplots_adjust(wspace = 0.5, hspace = 0.3)
plt.plot(x, 'r-', label='Tumour cells, T0=3')
plt.plot(y, 'b-', label='Hunting predator cell, H0=2')
plt.plot(z, 'g-', label='Resting predator cell, R0=1')
plt.title("Dynamics in time")
plt.xlabel("time")
plt.grid()
plt.legend(loc='best')
```
